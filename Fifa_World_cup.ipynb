Analyze Player-Level Data

```python
print(world_cup_players.describe())

# Top scorers in World Cup history
top_scorers = world_cup_players['Player Name'].value_counts().head(10)
print(top_scorers)

# Plot the top scorers
top_scorers.plot(kind='bar')
plt.title('Top Scorers in World Cup History')
plt.xlabel('Player Name')
plt.ylabel('Goals Scored')
plt.show()
```

Combine Datasets for Comprehensive Analysis

```python
# Extract 'Year' from 'Datetime' in world_cup_matches dataset
world_cup_matches['Year'] = pd.DatetimeIndex(world_cup_matches['Datetime']).year


# Merge World Cups and world_cup_matches datasets on 'Year'
combined_data = pd.merge(world_cup_matches, world_cups, on='Year', suffixes=('_match', '_cup'))

# Example analysis: Goals scored by the winning team in each match
goals_scored = combined_data.groupby('Winner')['GoalsScored'].sum()
print(goals_scored)

# Plot goals scored by winning teams
goals_scored.plot(kind='bar')
plt.title('Goals Scored by Winning Teams')
plt.xlabel('Team')
plt.ylabel('Total Goals')
plt.show()
```

Goals Scored Analysis


```python
# Total goals scored by each team in the World Cups
home_goals = combined_data.groupby('Home Team Name')['Home Team Goals'].sum()
away_goals = combined_data.groupby('Away Team Name')['Away Team Goals'].sum()
total_goals = home_goals.add(away_goals, fill_value=0).sort_values(ascending=False)
print(total_goals)

# Plot total goals scored by each team
total_goals.plot(kind='bar', figsize=(15, 7))
plt.title('Total Goals Scored by Each Team in World Cups')
plt.xlabel('Team')
plt.ylabel('Total Goals')
plt.show()
```

Attendance Analysis

```python
# Average attendance per match
avg_attendance = combined_data['Attendance_match'].mean()
print(f'Average Attendance per Match: {avg_attendance}')

# Total attendance for each World Cup
total_attendance_per_cup = combined_data.groupby('Year')['Attendance_match'].sum()
print(total_attendance_per_cup)

# Plot total attendance for each World Cup
total_attendance_per_cup.plot(kind='bar', figsize=(15, 7))
plt.title('Total Attendance for Each World Cup')
plt.xlabel('Year')
plt.ylabel('Total Attendance')
plt.show()
```

Match Outcomes Analysis

```python
# Win rates for teams
home_wins = combined_data[combined_data['Home Team Goals'] > combined_data['Away Team Goals']].groupby('Home Team Name').size()
away_wins = combined_data[combined_data['Away Team Goals'] > combined_data['Home Team Goals']].groupby('Away Team Name').size()
total_wins = home_wins.add(away_wins, fill_value=0).sort_values(ascending=False)
print(total_wins)

# Plot win rates for teams
total_wins.plot(kind='bar', figsize=(15, 7))
plt.title('Win Rates for Teams in World Cups')
plt.xlabel('Team')
plt.ylabel('Total Wins')
plt.show()
```

Combining Data for Comprehensive Insights

```python
# Example analysis: Goals scored by the winning team in each match
goals_by_winners_home = combined_data[combined_data['Home Team Name'] == combined_data['Winner']].groupby('Winner')['Home Team Goals'].sum()
goals_by_winners_away = combined_data[combined_data['Away Team Name'] == combined_data['Winner']].groupby('Winner')['Away Team Goals'].sum()
goals_by_winners = goals_by_winners_home.add(goals_by_winners_away, fill_value=0)
print(goals_by_winners)

# Plot goals scored by winning teams
goals_by_winners.plot(kind='bar', figsize=(15, 7))
plt.title('Goals Scored by Winning Teams')
plt.xlabel('Team')
plt.ylabel('Total Goals')
plt.show()
```
